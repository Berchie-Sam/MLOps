```python
def read_dataframe(filename):
    df = pd.read_csv(filename)
    df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime) 
    df.lpep_pickup_datetime  = pd.to_datetime(df.lpep_pickup_datetime)

    df['duration'] = df.lpep_dropoff_datetime - df.lpep_pickup_datetime
    df.duration = df.duration.apply(lambda td: td.total_seconds() / 60)

    df = df[(df.duration >= 0) & (df.duration <= 60)] 

    categorical = ['PULocationID', 'DOLocationID']
    df.loc[:, categorical] = df[categorical].astype(str)
    
    return df
```

```python
df_train = read_dataframe('D:\RGT\Code\MLOps\data\green_tripdata_2021-01.csv')
df_val = read_dataframe('D:\RGT\Code\MLOps\data\green_tripdata_2021-02.csv')
df_test = read_dataframe('D:\RGT\Code\MLOps\data\green_tripdata_2021-07.csv')
```

```python
def add_combined_column(df, c1, c2, new_column_name):
    df[new_column_name] = df[c1].astype(str) + '_' + df[c2].astype(str)
```

```python
add_combined_column(df_train, 'PULocationID', 'DOLocationID', 'PU_DO')
add_combined_column(df_val, 'PULocationID', 'DOLocationID', 'PU_DO')
add_combined_column(df_test, 'PULocationID', 'DOLocationID', 'PU_DO')
```

```python
df_train
```

```python
categorical =  ['PU_DO'] #['PULocationID', 'DOLocationID']
numerical = ['trip_distance']

dv = DictVectorizer()

train_dicts = df_train[categorical + numerical].to_dict(orient='records')
X_train = dv.fit_transform(train_dicts)

val_dicts = df_val[categorical + numerical].to_dict(orient='records')
X_val = dv.transform(val_dicts)

test_dicts = df_test[categorical + numerical].to_dict(orient='records')
X_test = dv.transform(test_dicts)
```

```python
X_train, X_val, X_test
```

```python
target = 'duration'
y_train = df_train[target].values
y_val = df_val[target].values
y_test = df_test[target].values
```

```python
with mlflow.start_run():
    mlflow.set_tag("developer", "Berchie")
    
    mlflow.log_param("train-data-path", "D:\RGT\Code\MLOps\data\green_tripdata_2021-01.csv")
    mlflow.log_param("val-data-path", "D:\RGT\Code\MLOps\data\green_tripdata_2021-02.csv")
    mlflow.log_param("test-data-path", "D:\RGT\Code\MLOps\data\green_tripdata_2021-07.csv")
    
    lr = LinearRegression()
    lr.fit(X_train, y_train)

    y_pred_lr =  lr.predict(X_val)

    mse_lr = mean_squared_error(y_val, y_pred_lr, squared=False)
    mlflow.log_metric("mse_lr", mse_lr)
    print(f'{mse_lr:.4f}')
    
    model_filename = "linear_regression_model.bin"
    model_filepath = "D:/RGT/Code/MLOps/models/" + model_filename
    
    with open(model_filepath, 'wb') as lr_model_file:
        pickle.dump((dv, lr), lr_model_file)
    
    mlflow.log_artifact(local_path=model_filepath, artifact_path="models_pickle")
```

```python
with mlflow.start_run():
    mlflow.set_tag("developer", "Berchie")
    
    mlflow.log_param("train-data-path", "D:\RGT\Code\MLOps\data\green_tripdata_2021-01.csv")
    mlflow.log_param("val-data-path", "D:\RGT\Code\MLOps\data\green_tripdata_2021-02.csv")
    mlflow.log_param("test-data-path", "D:\RGT\Code\MLOps\data\green_tripdata_2021-07.csv")
    
    alpha = 0.01
    mlflow.log_param("alpha", alpha)
    
    lasso = Lasso(alpha)
    lasso.fit(X_train, y_train)

    y_pred_lasso =  lasso.predict(X_val)

    mse_lasso = mean_squared_error(y_val, y_pred_lasso, squared=False)
    mlflow.log_metric("mse_lasso", mse_lasso)
    print(f'{mse_lasso:.4f}')
    
    model_filename = "lasso_regression_model.bin"
    model_filepath = "D:/RGT/Code/MLOps/models/" + model_filename
    
    with open(model_filepath, 'wb') as lasso_model_file:
        pickle.dump((dv, lasso), lasso_model_file)
    
    mlflow.log_artifact(local_path=model_filepath, artifact_path="models_pickle")
```

```python
with mlflow.start_run():
    mlflow.set_tag("developer", "Berchie")
    
    mlflow.log_param("train-data-path", "D:\RGT\Code\MLOps\data\green_tripdata_2021-01.csv")
    mlflow.log_param("val-data-path", "D:\RGT\Code\MLOps\data\green_tripdata_2021-02.csv")
    mlflow.log_param("test-data-path", "D:\RGT\Code\MLOps\data\green_tripdata_2021-07.csv")
    
    alpha = 0.001
    mlflow.log_param("alpha", alpha)

    ridge = Ridge(alpha)
    ridge.fit(X_train, y_train)

    y_pred_ridge = ridge.predict(X_val)

    mse_ridge = mean_squared_error(y_val, y_pred_ridge, squared=False)
    mlflow.log_metric("mse_ridge", mse_ridge)    
    print(f'{mse_ridge:.4f}')
    
    model_filename = "ridge_regression_model.bin"
    model_filepath = "D:/RGT/Code/MLOps/models/" + model_filename
    
    with open(model_filepath, 'wb') as ridge_model_file:
        pickle.dump((dv, ridge), ridge_model_file)

    mlflow.log_artifact(local_path="D:/RGT/Code/MLOps/models/ridge_regression_model.bin", artifact_path="models_pickle")
```
